Упр.1 : подключаем библиотеку, пишем необходимую функцию. Создаём точки для построения графика (список $x$). Подписываем оси и формулу функции в заголовке.

```python
import matplotlib.pyplot as plt

def ex2_m1 (x):
  return x**2*np.sin(1/x**2)+1

x = np.linspace(-1,1,250)
plt.plot(x,ex2_m1(x))
plt.xlabel(r'Ось $x$', fontsize=14)
plt.ylabel(r'Ось $y$', fontsize=14)
plt.title(r'$f(x)=x^2\sin(\dfrac{1}{x^2})+x$')
plt.axis('equal')
plt.show()
```

Упр. 3 : построим графики следующих функций на интервале $[0;1)$

* первая эллиптическая ф-я Якоби $sn$ с параметром 10;
* функция Бесселя первого рода порядка $1$, сжатая в $10$ раз;
* Кумулятивная функция распределения для 3х занчений при 10 испытаниях;
* Обратная зависимость среднего значения нормального распределения

```python
from scipy.special import ellipj, j1, bdtr, ndtri

x = np.arange(0,1,0.01)

plt.plot(x,ellipj(10,x)[0],label = r'$y = sn(x)$')
plt.plot(x,j1(x*10),label = r'$y=J_1(x)$')
plt.plot(x,bdtr(3,10,x),label = r'$y=\sum_{k=0}^{3}C_n^k x^k(1-x)^k$')
plt.plot(x,ndtri(x),label = r'$y=\mu^{-1}$')

plt.xlabel(r'Ось $x$', fontsize=10)
plt.ylabel(r'Ось $y$', fontsize=10)
plt.legend(loc='best', fontsize=10)
plt.show()
```

Упр. 4 : сгенерируем квадратную матрицу $A$, посчитаем $B=A^TA, C=AA^T$.

* Диагональные элементы считаем функцией $trace$.
* Определитель считаем функцией $linalg.det$.
* Ранг считаем функцией $linalg.matrix\_rank$.
* Спектр считаем функцией $linalg.eig$. При этом, отсекаем часть результата с собственными векторами.

В конце делаем выводы.

```python
n = int(input('Введите размер матрицы A >> '))
A = np.random.randint(10,size=(n,n))
print('\nМатрица A\n',A)
At = A.T
print('\nМатрица At\n',At)
B = At.dot(A)
print('\nМатрица B\n',B)
C = A.dot(At)
print('\nМатрица C\n',C)

print('\nСлед матрицы: tr(B) = ',B.trace(),
      '; tr(C) = ',C.trace())
print('Определитель: det(A) = {:.2f}'.format(np.linalg.det(A)),
      '; det(B) = {:.2f}'.format(np.linalg.det(B)),
      '; det(C) = {:.2f}'.format(np.linalg.det(C)))
print('Ранг матриц: rk(A) = ',np.linalg.matrix_rank(A),
      '; rk(B) = ',np.linalg.matrix_rank(B),
      '; rk(C) = ',np.linalg.matrix_rank(C))
print('Спектр B >>',np.linalg.eig(B)[0],
      '\nСпектр C >>',np.linalg.eig(C)[0])

print('\nВыводы:\n  1) матрицы B и C того же размера, что и A;',
      '\n  2) матрицы B и C симметрические;',
      '\n  3) определители B и C равны квадрату определителя A',
      '\n  4) ранги матриц A, B и C равны',
      '\n  5) спектры B и C совпадают.')
```

Выпал номер $2$. Напишем необходимую функцию и запомним крайние точки и погрешность.
Также посчитаем интеграл аналитически.

$\int\limits_1^2 \dfrac{e^{1/x^2}}{x^3}dx = (t = x^{-2}, dt = -2x^{-3}dx) \int\limits_1^{2^{-2}} \dfrac{e^tdt}{-2} = \dfrac{1}{2}\int_{1/4}^1e^tdt = \dfrac{1}{2}(e-\sqrt[4]{e})$.

Вычислим это значение уже с помощбю $python$.

```python
def f(x):
  return np.exp(1/x**2)/x**3

a = 1
b = 2
eps = 1e-5
int_f = 1/2*(np.exp(1)-np.exp(1)**0.25)
print('Интеграл = {:.5f}.'.format(int_f))
```

Посчитаем интеграл указанными методами. Для этого
* напишем функцию $equal\_$ для понимания, что три величины почти равны;
* введём счётчик $k$ количество итераций, а также счётчики $l1,l2,l3$ для понимания, как быстро сработал тот или иной метод;
* напишем векторизированную функцию $vect$ для работы со списком значений $x$;
* запустим цикл для изменения шага $h$ интегрирования;
* разметим отрезок $[a; b]$ с шагом $h$;
* посчитаем значения трёх интегралов.

```python
def equal_(a,b,c):
  #Функция проверки чисел на одинаковость
  return abs(a-b)<eps and abs(b-c)<eps and abs(c-a)<eps

h = 1
k = 0
l1,l2,l3 = 0,0,0
f1,f2,f3 = 0,1,2
vect = np.vectorize(lambda t: f(t))
while not equal_(f1,f2,f3):
  h /= 10
  k += 1
  x = np.arange(a,b+h,h)
  vect
  f1 = h * np.sum(vect(x))
  f2 = h/2 * (f(a) + f(b) + 2*np.sum(vect(x[1:-1])))
  f3 = h/3 * (f(a) + f(b) + 2*np.sum(vect(x[2:-1:2])) + 4*np.sum(vect(x[1:-1:2])))
  if abs(f1-int_f) < eps and l1 == 0: l1 = k
  if abs(f2-int_f) < eps and l2 == 0: l2 = k
  if abs(f3-int_f) < eps and l3 == 0: l3 = k

print('Проведено',k,'итераций.')
print('Найденные значения:')
print('методом прямоугольников >  f1 = {:.6f}'.format(f1),'  за',l1,'операций ')
print('методом трапеций        >  f2 = {:.6f}'.format(f2),'  за',l2,'операций ')
print('по правилу Симпсона     >  f3 = {:.6f}'.format(f3),'  за',l3,'операций ')
```
